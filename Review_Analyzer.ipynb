```python
data_path = "/content/new_updated_review.json"
with open(data_path, "r") as f:
    data = json.load(f)
```

```python
df= pd.json_normalize(data)
df.head()
```

DATA NORMALIZATION

```python
df['stars'] = pd.to_numeric(df['stars'], errors='coerce')
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['author'] = df['author'].fillna("Unknown")
df['company'] = df['company'].fillna("Unknown")
df['company_size'] = df['company_size'].fillna("Unknown")
df['categories'] = df['categories'].apply(lambda x: ", ".join(x) if isinstance(x, list) else x)
df['ratings'] = df['ratings'].apply(lambda r: r if isinstance(r, list) else [])
```

DATA SORTING FOR DROPDOWN OPTIONS

```python
author_list = sorted(df["author"].unique())
date_list = sorted(df["date"].dropna().astype(str).unique())
company_list = sorted(df["company"].unique())
stars_list = sorted(df["stars"].dropna().unique().astype(str))
category_list = sorted(set(cat for sublist in df["categories"].dropna().str.split(", ") for cat in sublist))
company_size_list = sorted(df["company_size"].unique())
rating_list = sorted(set(r for ratings in df["ratings"] for r in ratings if isinstance(r, (int, float))))
```

ANALYSIS GRAPHS

```python
def fig_to_base64(fig):
    buf = io.BytesIO()
    fig.savefig(buf, format="png", bbox_inches='tight')
    buf.seek(0)
    img_b64 = base64.b64encode(buf.read()).decode("utf-8")
    plt.close(fig)
    return img_b64
```

```python
def generate_graphs(filtered_df):
    graphs = []

    if not filtered_df.empty:
        try:
            category_counts = filtered_df['categories'].str.split(', ', expand=True).stack().value_counts().reset_index()
            category_counts.columns = ['Category', 'Count']
            fig2 = plt.figure(figsize=(10, 7))
            sns.barplot(data=category_counts, x='Count', y='Category', palette="Set2")
            plt.title("Reviews by Category")
            graphs.append(fig_to_base64(fig2))
        except:
            graphs.append(None)
    else:
        graphs.append(None)

    if not filtered_df.empty:
        try:
            exploded = filtered_df.copy()
            exploded['Category'] = exploded['categories'].str.split(', ')
            exploded = exploded.explode('Category')
            exploded = exploded.explode('ratings')
            exploded = exploded.dropna(subset=['Category', 'ratings'])
            exploded['ratings'] = pd.to_numeric(exploded['ratings'], errors='coerce')
            rating_order = [1, 2, 3, 4, 5]
            heatmap_data = pd.crosstab(exploded['Category'], exploded['ratings'])[rating_order].fillna(0)
            fig3 = plt.figure(figsize=(14, max(8, len(heatmap_data) * 0.4)))
            sns.heatmap(heatmap_data, annot=True, fmt='g', cmap="YlGnBu", linewidths=0.5, cbar_kws={'label': 'Count'})
            plt.title("Ratings vs Categories Heatmap")
            plt.xlabel("Rating")
            plt.ylabel("Category")
            plt.tight_layout()
            graphs.append(fig_to_base64(fig3))
        except:
            graphs.append(None)
    else:
        graphs.append(None)

    if not filtered_df.empty:
        counts = filtered_df['company_size'].value_counts()
        if not counts.empty:
            fig4 = plt.figure(figsize=(6, 6))
            plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
            plt.title("Company Size Distribution")
            plt.axis('equal')
            graphs.append(fig_to_base64(fig4))
        else:
            graphs.append(None)
    else:
        graphs.append(None)

    if not filtered_df.empty and 'date' in filtered_df:
        try:
            timeline = filtered_df['date'].dropna().dt.to_period("M").value_counts().sort_index().reset_index()
            timeline.columns = ['Month', 'Count']
            timeline['Month'] = timeline['Month'].astype(str)
            fig5 = plt.figure(figsize=(12, 6))
            sns.lineplot(data=timeline, x='Month', y='Count', marker='o')
            plt.title("Reviews Over Time")
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            graphs.append(fig_to_base64(fig5))
        except:
            graphs.append(None)
    else:
        graphs.append(None)

    return graphs
```

DROPDOWN FILTERS

```python
def filter_data(authors, dates, stars, companies, categories, company_sizes, ratings):
    filtered = df.copy()
    if authors:
        filtered = filtered[filtered["author"].isin(authors)]
    if dates:
        filtered = filtered[filtered["date"].astype(str).isin(dates)]
    if stars:
        filtered = filtered[filtered["stars"].astype(str).isin(stars)]
    if companies:
        filtered = filtered[filtered["company"].isin(companies)]
    if categories:
        filtered = filtered[filtered["categories"].apply(lambda x: any(cat in x for cat in categories))]
    if company_sizes:
        filtered = filtered[filtered["company_size"].isin(company_sizes)]
    if ratings:
        filtered = filtered[filtered["ratings"].apply(lambda x: any(float(r) in x for r in ratings))]

    graphs = generate_graphs(filtered)
    return filtered, graphs
```

GRADIO INTERFACE

```python
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("## 📊 Review Analytics Dashboard")
    gr.Markdown("Use the filters below to explore visual insights.")

    with gr.Row():
        author_dropdown = gr.Dropdown(choices=author_list, multiselect=True, label="Author(s)")
        date_dropdown = gr.Dropdown(choices=date_list, multiselect=True, label="Date(s)")

    with gr.Row():
        stars_dropdown = gr.Dropdown(choices=stars_list, multiselect=True, label="Stars")
        company_dropdown = gr.Dropdown(choices=company_list, multiselect=True, label="Company")
        category_dropdown = gr.Dropdown(choices=category_list, multiselect=True, label="Category")

    with gr.Row():
        company_size_dropdown = gr.Dropdown(choices=company_size_list, multiselect=True, label="Company Size")
        rating_dropdown = gr.Dropdown(choices=[str(r) for r in rating_list], multiselect=True, label="Ratings")

    # Apply wrap to text cells
    output_table = gr.Dataframe(label="Filtered Results", wrap=True)

    graph1 = gr.Image(label="Rating Distribution")
    graph2 = gr.Image(label="Reviews by Category")
    graph3 = gr.Image(label="Ratings vs Categories")
    graph4 = gr.Image(label="Company Size Distribution")
    graph5 = gr.Image(label="Reviews Over Time")

    inputs = [author_dropdown, date_dropdown, stars_dropdown,
              company_dropdown, category_dropdown, company_size_dropdown, rating_dropdown]

    def update_display(*args):
        filtered, graphs = filter_data(*args)
        images = [Image.open(BytesIO(base64.b64decode(g))) if g else None for g in graphs]
        return [filtered] + images

    for inp in inputs:
        inp.change(fn=update_display, inputs=inputs,
                   outputs=[output_table, graph1, graph2, graph3, graph4, graph5])

    demo.launch()
```
